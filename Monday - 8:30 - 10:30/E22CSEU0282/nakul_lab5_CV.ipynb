```python
import matplotlib.pyplot as plt
from PIL import Image
import os

def compress_and_show(image_path, output_folder="compressed_images", jpeg_quality=50):
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Load image
    img = Image.open(image_path)

    # Lossy compression (JPEG)
    jpeg_output_path = os.path.join(output_folder, "compressed_image.jpeg")
    img.save(jpeg_output_path, "JPEG", quality=jpeg_quality)

    # Lossless compression (PNG)
    png_output_path = os.path.join(output_folder, "compressed_image.png")
    img.save(png_output_path, "PNG", optimize=True)

    # Load compressed images
    img_jpeg = Image.open(jpeg_output_path)
    img_png = Image.open(png_output_path)

    # Display original and compressed images
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))

    axes[0].imshow(img)
    axes[0].set_title("Original Image")
    axes[0].axis("off")

    axes[1].imshow(img_jpeg)
    axes[1].set_title("Lossy JPEG Compression")
    axes[1].axis("off")

    axes[2].imshow(img_png)
    axes[2].set_title("Lossless PNG Compression")
    axes[2].axis("off")

    plt.show()

# Example usage
image_path = "/content/Parrot.jpg"
compress_and_show(image_path)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from keras.datasets import mnist, cifar10
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Function to preprocess data
def preprocess_data(x_train, x_test, y_train, y_test, num_classes):
    x_train, x_test = x_train / 255.0, x_test / 255.0  # Normalize
    y_train = keras.utils.to_categorical(y_train, num_classes)
    y_test = keras.utils.to_categorical(y_test, num_classes)
    return x_train, x_test, y_train, y_test

# Function to build a CNN model
def build_cnn(input_shape, num_classes):
    model = keras.Sequential([
        layers.Conv2D(32, kernel_size=(3, 3), activation="relu", input_shape=input_shape),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dense(128, activation="relu"),
        layers.Dense(num_classes, activation="softmax")
    ])
    model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
    return model

# Function to train and evaluate CNN
def train_and_evaluate(dataset_name="mnist"):
    print(f"\n Training CNN on {dataset_name.upper()} dataset...\n")

    # Load dataset
    if dataset_name == "mnist":
        (x_train, y_train), (x_test, y_test) = mnist.load_data()
        x_train = x_train.reshape(-1, 28, 28, 1)  # Reshape for CNN
        x_test = x_test.reshape(-1, 28, 28, 1)
        num_classes = 10
    elif dataset_name == "cifar10":
        (x_train, y_train), (x_test, y_test) = cifar10.load_data()
        num_classes = 10

    # Preprocess data
    x_train, x_test, y_train, y_test = preprocess_data(x_train, x_test, y_train, y_test, num_classes)

    # Train-test split (80-20)
    split = int(0.8 * len(x_train))
    x_train, x_val = x_train[:split], x_train[split:]
    y_train, y_val = y_train[:split], y_train[split:]

    # Build and train CNN model
    input_shape = x_train.shape[1:]  # (28,28,1) for MNIST, (32,32,3) for CIFAR-10
    model = build_cnn(input_shape, num_classes)
    history = model.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=50, batch_size=64, verbose=2)

    # Evaluate model
    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
    print(f"\n Test Accuracy: {test_acc:.4f}")

    # Predictions
    y_pred = model.predict(x_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_test_classes = np.argmax(y_test, axis=1)

    # Compute metrics
    report = classification_report(y_test_classes, y_pred_classes, output_dict=True)
    print("\n Classification Report:\n", classification_report(y_test_classes, y_pred_classes))

    precision = report["weighted avg"]["precision"]
    recall = report["weighted avg"]["recall"]
    f1_score = report["weighted avg"]["f1-score"]

    print(f" Precision: {precision:.4f}")
    print(f" Recall: {recall:.4f}")
    print(f" F1-Score: {f1_score:.4f}")

    # Confusion Matrix
    cm = confusion_matrix(y_test_classes, y_pred_classes)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=range(10), yticklabels=range(10))
    plt.title(f"{dataset_name.upper()} Confusion Matrix")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.show()

    # ROC & AUC
    fpr, tpr, _ = roc_curve(y_test.ravel(), y_pred.ravel())
    roc_auc = auc(fpr, tpr)
    print(f" AUC Score: {roc_auc:.4f}")

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color="blue", label=f"AUC = {roc_auc:.2f}")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"{dataset_name.upper()} ROC Curve")
    plt.legend()
    plt.show()

# Run CNN for MNIST
train_and_evaluate("mnist")

# Run CNN for CIFAR-10
train_and_evaluate("cifar10")
```
