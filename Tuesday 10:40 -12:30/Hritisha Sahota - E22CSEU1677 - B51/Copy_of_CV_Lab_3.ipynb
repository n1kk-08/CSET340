```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Read the image
image_bgr = cv2.imread('/content/image_2.16.12.jpeg')

# 2. Display the image
plt.imshow(cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

# 3. Extraction of Image size
height, width, channels = image_bgr.shape
print(f"Image Size: {width}x{height}, Channels: {channels}")

# 4. Calculation of image pixels
total_pixels = height * width
print(f"Total Pixels: {total_pixels}")

# 5. Convert RGB to Grayscale image
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

# Display Grayscale Image
plt.imshow(image_gray, cmap='gray')
plt.axis('off')
plt.show()

# 6. Convert RGB to Binary image using a threshold
_, image_binary = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY)

# Display Binary Image
plt.imshow(image_binary, cmap='gray')
plt.axis('off')
plt.show()

# Count the area of black pixels
black_pixels = np.sum(image_binary == 0)
print(f"Area of Black Pixels: {black_pixels}")
```

```python
# Task-2: Edge Detection
from google.colab.patches import cv2_imshow

# Sobel Operator
sobel_x = cv2.Sobel(image_gray, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(image_gray, cv2.CV_64F, 0, 1, ksize=5)
sobel_combined = cv2.magnitude(sobel_x, sobel_y)

# Display Sobel Edge Detection
plt.imshow(sobel_combined, cmap='gray')
plt.axis('off')
plt.title('Sobel Edge Detection')
plt.show()

# Prewitt Operator (using convolution)
prewitt_kernel_x = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
prewitt_kernel_y = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
prewitt_x = cv2.filter2D(image_gray.astype(np.float32), -1, prewitt_kernel_x)
prewitt_y = cv2.filter2D(image_gray.astype(np.float32), -1, prewitt_kernel_y)
prewitt_combined = cv2.magnitude(prewitt_x, prewitt_y)

# Display Prewitt Edge Detection
plt.imshow(prewitt_combined, cmap='gray')
plt.axis('off')
plt.title('Prewitt Edge Detection')
plt.show()

# Roberts Cross Operator
roberts_kernel_x = np.array([[1, 0], [0, -1]])
roberts_kernel_y = np.array([[0, 1], [-1, 0]])
roberts_x = cv2.filter2D(image_gray.astype(np.float32), -1, roberts_kernel_x)
roberts_y = cv2.filter2D(image_gray.astype(np.float32), -1, roberts_kernel_y)

# Check if the outputs are valid
roberts_combined = cv2.magnitude(roberts_x, roberts_y)

# Display Roberts Cross Edge Detection
plt.imshow(roberts_combined, cmap='gray')
plt.axis('off')
plt.title('Roberts Cross Edge Detection')
plt.show()

# Canny Edge Detector
edges_canny = cv2.Canny(image_gray, 100, 200)

# Display Canny Edge Detection
plt.imshow(edges_canny, cmap='gray')
plt.axis('off')
plt.title('Canny Edge Detection')
plt.show()

# Task-2: Image Segmentation

# Global Thresholding
```
